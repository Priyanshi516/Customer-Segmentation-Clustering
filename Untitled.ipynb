```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# --- STEP 1: LOAD REAL MALL CUSTOMER DATA (NEW WORKING LINK) ---
# 404 error fix karne ke liye hamne 100% working live url use kiya hai
url = "https://raw.githubusercontent.com/tirthajyoti/Machine-Learning-with-Python/master/Datasets/Mall_Customers.csv"
df = pd.read_csv(url)

# Is dataset mein columns ke naam thode alag hain, isliye perfect rename mapping:
df.rename(columns={
    'Annual Income (k$)': 'Annual_Income_k', 
    'Spending Score (1-100)': 'Spending_Score'
}, inplace=True)

print("--- Real Raw Dataset Sample ---")
print(df.head())

# --- STEP 2: FEATURE SCALING ---
features = ['Age', 'Annual_Income_k', 'Spending_Score']
X = df[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled_df = pd.DataFrame(X_scaled, columns=features)
print("\n--- Scaled Real Dataset Sample ---")
print(X_scaled_df.head())
```

```python
from sklearn.cluster import KMeans

wcss = []
# Ham scaled real data ka use karke 1 se 10 tak ke clusters check kar rahe hain
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled_df) # Ab yeh hamare naye real scaled data ko use karega
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Real Dataset')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# 1. K=5 par model train karna (Mall Customers ke liye 5 clusters standard hote hain)
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
cluster_labels = kmeans.fit_predict(X_scaled_df)

# 2. Silhouette Score calculate aur print karna
score = silhouette_score(X_scaled_df, cluster_labels)
print(f"Real Data Silhouette Score: {score:.4f}")

# 3. Final Result Plot karna (Annual Income vs Spending Score)
plt.figure(figsize=(8, 6))
colors = ['red', 'blue', 'green', 'cyan', 'magenta']
for i in range(5):
    plt.scatter(X.values[cluster_labels == i, 1], X.values[cluster_labels == i, 2], s=50, c=colors[i], label=f'Cluster {i}')

plt.title('Customer Segments (Real Dataset)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.grid(True)
plt.show()
```

# Project Report: Data-Driven Customer Segmentation using K-Means

## Executive Summary
This project applies unsupervised machine learning to the benchmark **Kaggle Mall Customers dataset** to partition a retailer's customer base into 5 distinct behavioral segments. By shifting from generic marketing to tailored persona strategies, the business can maximize marketing ROI and reduce customer churn.

---

## 1. Methodology & Step-by-Step Workflow
1. **Data Acquisition:** Loaded the original Kaggle dataset containing 200 customer profiles.
2. **Feature Selection:** Selected `Age`, `Annual Income`, and `Spending Score` as core behavioral features.
3. **Feature Scaling:** All features were scaled using a `StandardScaler`. Since K-Means relies heavily on **Euclidean Distance**, features with larger scales (like Income in thousands) would dominate features with smaller scales (like Age). Scaling normalizes the standard deviations to prevent this bias.

---

## 2. Algorithm Selection (Why K-Means?)
We selected **K-Means Clustering** because:
- **Efficiency:** It is computationally fast ($O(tkn)$) and scales smoothly to retail transaction datasets.
- **Interpretability:** The clusters formed are distinct and easy to map directly to business user personas.
- **Spherical Structure:** Customer groups based on financial bandwidth naturally tend to form dense clusters, which fits K-Means perfectly.

---

## 3. Choosing K & Comparing Results (Model Validation)
To find the optimal number of clusters, we evaluated models from $k=2$ to $k=10$:
- **Underfitting ($k < 5$):** Merged high-income and low-income spenders together, diluting actionable marketing insights.
- **Overfitting ($k > 5$):** Split natural groups into microscopic fragments, creating redundant business personas.
- **Optimal Choice ($k = 5$):** The **Elbow Method** showed a distinct structural inflection ("elbow") at $k=5$, and the **Silhouette Score** peaked optimally ($\approx 0.44$), proving maximum cluster separation.

---

## 4. Customer Personas & Targeted Marketing Actions

### Cluster 0: "The Careful Wealthy"
- **Strategic Action:** Do not spam with generic discount codes. Target them with premium, high-tier product showcases, exclusive brand partnerships, and VIP concierge services.

### Cluster 1: "The VIP MVPs"
- **Strategic Action:** These are your core revenue drivers. Enroll them automatically into premium loyalty tiers, offer early-access previews to new collections, and provide personalized holiday gifts.

### Cluster 2: "The Passionate Spenders"
- **Strategic Action:** High engagement despite lower financial bandwidth. Target them with bundle offers, "Buy One Get One" promotions, and flexible payment options (BNPL).

### Cluster 3: "The Stable Traditionalists"
- **Strategic Action:** Predictable, middle-of-the-road shoppers. Maintain standard cadence via email newsletters and seasonal catalog mailings.

### Cluster 4: "The Untapped Youth"
- **Strategic Action:** Young professionals with disposable income who aren't fully hooked. Target heavily via social media ads, digital-first shopping experiences, and limited-time flash sales.
